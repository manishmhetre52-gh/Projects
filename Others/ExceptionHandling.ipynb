```python
"""1. Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero."""
def div(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return"Division by zero is not possible"
```

```python
print(div(1,2))
print(div(2,0))
```

```python
"""2. Write a Python program that prompts the user to input an integer and raises a ValueError exception if the input is not a valid integer."""
def input_int():
    check = input("Please enter an integer: ")
    try:
        int(check)
        return "Thanks for the integer input"
    except ValueError:
        return "You did not enter an integer"
```

```python
input_int()
```

```python
"""3. Write a Python program that opens a file and handles a FileNotFoundError exception if the file does not exist."""
def open_this(file_name):
    try:
        f = open(file_name)
        return "Yes this file exists"
    except FileNotFoundError:
        return "File does not exist"
```

```python
open_this("exercise4.txt")
```

```python
"""4. Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical."""
def input_two():
    a = input("Enter integer no 1:").strip()
    b = input("Enter integer no 2:").strip()
    try:
        int(a)
        int(b)
        return "Thanks for the integer input"
    except ValueError:
        raise TypeError("Both of the inputs are not integers")
```

```python
input_two()
```

```python
"""5. Write a Python program that opens a file and handles a PermissionError exception if there is a permission issue."""
os.chmod("pass.txt",0o000) #changing permission settings
```

```python
def write_this_file(file_name,string):
    try:
        with open(file_name,"w") as file:
            file.write(string)
    except PermissionError as e:
        return e,"You do not have permission to write the file"
```

```python
write_this_file("pass.txt","asdfghjkl")
```

```python
os.chmod("pass.txt",0o644) #permission settings back to default
```

```python
with open("pass.txt",'r') as file:
    print(file.read())
```
