```python
"""5. Write a Python program to create a class representing a binary search tree. Include methods for inserting and searching for elements in the binary tree."""
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.head = None
    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node
    def display(self):
        if self.head is None:
            print("Linked List is empty")
            return
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")
    def delete(self, value):
        current = self.head
        if current and current.data == value:
            self.head = current.next
            return f"{value} deleted successfully"
        prev = None
        while current and current.data != value:
            prev = current
            current = current.next
        if current is None:
            return f"{value} not found"
        prev.next = current.next
        return f"{value} deleted successfully"
```

```python
"""7. Write a Python program to create a class representing a linked list data structure. Include methods for displaying linked list data, inserting and deleting nodes.
"""
class Node:
    def __init__(self,value):
        self.right = None
        self.left = None
        self.value = value
    def insert(self,value):
        if value < self.value:
            if self.left is None:
                self.left = Node(value)
            else:
                self.left.insert(value)
        else:
            if self.right is None:
                self.right = Node(value)
            else:
                self.right.insert(value)
    def delete(self, value):
        if value < self.value:
            if self.left:
                self.left = self.left.delete(value)
        elif value > self.value:
            if self.right:
                self.right = self.right.delete(value)
        else:
            if self.left is None:
                return self.right
            if self.right is None:
                return self.left
            temp = self.right
            while temp.left:
                temp = temp.left
            self.value = temp.value
            self.right = self.right.delete(temp.value)
        return self
```

```python
"""8. Write a Python program to create a class representing a shopping cart. Include methods for adding and removing items, and calculating the total price. """
class ShoppingCart:
    def __init__(self):
        self.items = []
    def add_items(self,product,price):
        self.items.append([product,price])
        return f"{product.capitalize()} has been added to the shopping cart."
    def remove_items(self,product):
        product = product.lower()
        for item in self.items:
            if item[0].lower() == product:
                self.items.remove(item)
                return f"{product.capitalize()} has been removed from the shopping cart."
        return f"{product.capitalize()} does not exist in the shopping cart."
    def calc_total(self):
        total = 0
        for item in self.items:
            total += item[1]
        return f"Total price of your shopping cart is {total}₹ "
```

```python
"""14. Write a program that uses datetime module within a class. Enter manufacturing date and expiry date of the product. The program must display the years, months and days that are left for expiry."""
class Storage:
    def __init__(self,product_name,manufacturing_date,expiry_date):
        self.product_name = product_name
        self.manufacturing_date = manufacturing_date
        self.expiry_date = expiry_date
    def time_till_expiry(self):
        today = date.today()
        day,month,year = self.expiry_date
        expiry = date(year,month,day)
        if expiry < today:
            return f"Your product {self.product_name} has expired"
        remaining = relativedelta(expiry,today)
        return f"Product {self.product_name}: remaining years {remaining.years} months {remaining.months} days {remaining.days}"
```

```python
"""18. Write a Python class Employee with attributes like emp_id, emp_name, emp_salary, and emp_department and methods like calculate_emp_salary, emp_assign_department, and print_employee_details.
	Sample Employee Data:
	"ADAMS", "E7876", 50000, "ACCOUNTING"
	"JONES", "E7499", 45000, "RESEARCH"
	"MARTIN", "E7900", 50000, "SALES"
	"SMITH", "E7698", 55000, "OPERATIONS"
	Use 'assign_department' method to change the department of an employee.
	Use 'print_employee_details' method to print the details of an employee.
	Use 'calculate_emp_salary' method takes two arguments: salary and hours_worked, which is the number of hours worked by the employee. If the number of hours worked is more than 50, the method computes overtime and adds it to the salary. Overtime is calculated as following formula:
	overtime = hours_worked - 50
	Overtime amount = (overtime * (salary / 50))
	"""
class Employee:
    def __init__(self,emp_name,emp_id,emp_salary,emp_department):
        self.emp_name = emp_name
        self.emp_id = emp_id
        self.emp_salary = emp_salary
        self.emp_department = emp_department
    def calculate_emp_salary(self,hours_worked):
        if hours_worked <= 50:
            salary = self.emp_salary
            return f"Salary for {self.emp_name} is Rs.{salary} "
        else:
            overtime = hours_worked - 50
            salary = self.emp_salary + (overtime * (self.emp_salary/ 50))
            return f"Salary for {self.emp_name} is Rs.{salary}"
    def assign_department(self,department):
        if self.emp_department == department.upper():
            return f"Employee {self.emp_name} is already assigned with department {self.emp_department}"
        else:
            self.emp_department = department.upper()
            return f"Employee {self.emp_name} is assigned with department {department} successfully"
    def print_employee_details(self):
        return self.emp_name,self.emp_id,self.emp_salary,self.emp_department
```

```python
"""19. Write a Python class Restaurant with attributes like menu_items, book_table, and customer_orders, and methods like add_item_to_menu, book_tables, and customer_order.
	Perform the following tasks now:
	Now add items to the menu.
	Make table reservations.
	Take customer orders.
	Print the menu.
	Print table reservations.
	Print customer orders.
	Note: Use dictionaries and lists to store the data."""
class Restaurant:
    reserved_table = None
    def __init__(self,tables):
        self.tables = list(range(1,tables+1))
        self.menu_list = {}
        self.customer_order_list = {}
    def show_menu(self):
        return self.menu_list
    def make_reservation(self):
        cust_name = input("Enter Customer Name: ").strip().lower()
        if cust_name not in self.menu_list.keys():
            if self.tables:
                self.reserved_table = random.choice(self.tables)
                self.tables.remove(self.reserved_table)
                self.customer_order_list[cust_name] = [self.reserved_table]
                return f"Table No. {self.reserved_table} is now reserved for customer {cust_name}"
            else:
                return "Sorry, Tables are not available to make a reservation"
    def add_to_menu(self):
        item = input("Enter Menu Item: ").strip().lower()
        price = int(input("Enter Price: "))
        if item in self.menu_list.keys():
            return f"Item {item} already exists in the menu"
        self.menu_list[item] = price
        return f"Item {item} added to menu"
    def update_price(self):
        item = input("Enter the item that you want to update: ").strip().lower()
        if item in self.menu_list.keys():
            set_price = int(input("Enter new Price: "))
            self.menu_list[item] = set_price
            return f"Price for {item} has been updated successfully to {set_price}"
        return f"Item {item} does not exist in the menu"
    def customer_order(self):
        cust_name = input("Enter Customer Name: ").strip().lower()
        if cust_name in self.customer_order_list:
            while True:
                item = input("Enter Menu Item('STOP' to end the order): ").strip().lower()
                if item == 'stop':
                    return f"Order has been executed successfully"
                if item in self.menu_list.keys():
                    self.customer_order_list[cust_name].append(item)
                    print(f"Item {item} added to customer order")
                else:
                    print(f"Item {item} does not exist in the menu")
        return f"Customer {cust_name} does not have a reservation"
    def print_bill(self):
        cust_name = input("Enter Customer Name: ").strip().lower()
        tot = 0
        if cust_name in self.customer_order_list.keys():
            for item in self.customer_order_list[cust_name][1::]:
                print(f"{item}: {self.menu_list[item]}")
                tot += self.menu_list[item]
            if self.customer_order_list[cust_name] is not None:
                self.tables.append(self.customer_order_list[cust_name][0])
                self.reserved_table = None
                del self.customer_order_list[cust_name]
            return f"Customer Name: {cust_name}\nTotal bill: {tot}"
        return f"Customer {cust_name} does not have a reservation"
    def main(self):
        while True:
            menu = input("Welcome to the Interface!\nChoose Your Option:\n1.Add an item to the menu\n2.Update Menu Price\n3.Make Reservation\n4.Take Customer Order\n5.Generate Bill\n6.Exit").strip().lower()
            if menu == "1":
                print(self.add_to_menu())
            elif menu == "3":
                print(self.make_reservation())
            elif menu == "2":
                print(self.update_price())
            elif menu == "4":
                print(self.customer_order())
            elif menu == "5":
                print(self.print_bill())
            elif menu == "6":
                print("Thank you for using this program")
                break
            else:
                print("Please enter a valid option")
```

```python
"""20. Write a Python class BankAccount with attributes like account_number, balance, date_of_opening and customer_name, and methods like deposit, withdraw, and check_balance."""
class BankAccount:
    def __init__(self,account_number,balance,date_of_opening,customer_name):
        self.account_number = account_number
        self.balance = balance
        self.date_of_opening = date_of_opening
        self.customer_name = customer_name
    def deposit(self,amount):
        if amount > 0:
            self.balance += amount
            return f"Rs.{amount} has been deposited . Your balance is Rs.{self.balance}"
        else:
            return "Invalid amount entered"
    def withdraw(self,amount):
        if amount <= self.balance:
            if amount > 0:
                if amount%100==0:
                    self.balance += amount
                    return f"Rs.{amount} has beem withdrawn from you account, Your balance is Rs.{self.balance}"
                else:
                    return "Invalid amount entered. Only Multiples of 100 are allowed"
            else:
                return "Withdraw Amount cannot be zero or below zero"
        else:
            return "Not Enough Balance"
    def check_balance(self):
        return f"Your account balance is Rs.{self.balance}"
```

```python
"""21. Write a Python class Inventory with attributes like item_id, item_name, stock_count, and price, and methods like add_item, update_item, and check_item_details."""
class Inventory:
    def __init__(self):
        self.item_dict = {}
    def add_item(self):
        id = int(input("Enter Item ID: "))
        if id in self.item_dict.keys():
            return f"Item {id} already exists in the inventory"
        else:
            name = input("Enter Item Name: ").strip().lower()
            count = int(input("Enter Item Count: "))
            price = int(input("Enter Item Price: "))
            self.item_dict[id] = [name,count,price]
            return f"Item {id} added to inventory"
    def update_item(self):
        id = int(input("Enter Item ID to update: "))
        if id in self.item_dict.keys():
            name = input("Enter Item New Name: ").strip().lower()
            count = int(input("Enter new Item Count: "))
            price = int(input("Enter Item Price: "))
            self.item_dict[id] = [name,count,price]
            return f"Details for {id} updated successfully"
        return f"Item {id} does not exist in the inventory"
    def check_item_details(self):
        id = int(input("Enter Item ID: "))
        if id in self.item_dict.keys():
            return f"Item ID: {id}, Item Name: {self.item_dict[id][0]}, Item Count:{self.item_dict[id][1]}, Item Price:{self.item_dict[id][2]}"
        return f"Item {id} does not exits in the inventory"
```
