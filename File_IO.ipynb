```python
"""2. Write a Python program to read first n lines of a file."""
def read_first_n_lines(n,file_name):
    file = open(file_name,'r')
    for i in range(n):
        print(file.readline())
    file.close()
```

```python
read_first_n_lines(10,"exercise.txt")
```

```python
"""3. Write a Python program to append text to a file and display the text."""
def append_text(string,filename):
    with open(filename,'a') as f:
        f.write(string)
    with open(filename,'r') as l:
        print(l.read())
```

```python
append_text("hello","exercise.txt")
```

```python
"""4. Write a Python program to read last n lines of a file."""
def read_last_n_lines(n,filename):
    with open("exercise.txt",'r') as f:
        as_list = f.read().split('\n')
        last = as_list[-1:-(n+1):-1]
        print(" \n".join(last))
```

```python
read_last_n_lines(10,"exercise.txt")
```

```python
"""7. Write a python program to find the longest words."""
def find_longest_word(filename):
    with open(filename,'r') as f:
        as_list = f.read().split()
        max_len = 0
        longest_word = ""
        for word in as_list:
            if len(word) > max_len:
                max_len = len(word)
                longest_word = word
        print(longest_word)
```

```python
find_longest_word("exercise.txt")
```

```python
"""8. Write a Python program to count the number of lines in a text file."""
def count_lines(filename):
    with open(filename,'r') as f:
        as_list = f.read().split("\n")
        print(len(as_list))
```

```python
count_lines("exercise.txt")
```

```python
"""9. Write a Python program to count the frequency of words in a file."""
def count_words(filename,words):
    count = 0
    with open(filename,'r') as f:
        read_file = f.read()
        for word in read_file.split():
            if word == words:
                count += 1
        return count
```

```python
count_words("exercise.txt","is")
```

```python
"""11. Write a Python program to write a list to a file."""
def list_in_file(filename,lst):
    with open(filename,'w') as f:
        json.dump(lst,f)
    print(f"{lst} has been added to {filename} succesfully")
```

```python
"""12. Write a Python program to copy the contents of a file to another file ."""
def copy_file_to_file(file1,file2):
    data = None
    with open(file1,"r") as f1:
        data = f1.read()
    with open(file2,"w") as f2:
        f2.write(data)
    print(f"Data successfully copied from {file1} to {file2}")
```

```python
copy_file_to_file("exercise.txt","exercise2.txt")
```

```python
"""13. Write a Python program to read a random line from a file."""
def read_random_line(filename):
    with open(filename,'r') as f:
        data = f.read().split("\n")
    pos = random.randint(0,len(data)-1)
    return str(data[pos])
```

```python
read_random_line("exercise.txt")
```

```python
"""14. Write a Python program to assess if a file is closed or not."""
def closed_or_not(filename):
    temp_file = open(filename,'r')
    print("Before closing:",temp_file.closed)
    temp_file.close()
    print("After closing:",temp_file.closed)
```

```python
closed_or_not("exercise.txt")
```

```python
"""15. Write a Python program that takes a text file as input and returns the number of words of a given text file."""
def count_tot_words(filename):
    with open(filename,'r') as f:
        data = f.read().split()
    return f"Total number of words: {len(data)}"
```

```python
count_tot_words("exercise.txt")
```
