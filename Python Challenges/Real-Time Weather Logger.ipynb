```python
class WeatherLogger:
    def __init__(self):
        self.log_file = "weather_log.csv"
        self.api_key = "2c425b16ec3e062286d6de7ecd47a82e"
    def get_weather_info(self,city):
        url = "https://api.openweathermap.org/data/2.5/weather"
        parameters = {"q": city, "appid": self.api_key,"units": "metric"}
        try:
            response = requests.get(url, params=parameters,timeout=5)
            data = response.json()
            if response.status_code != 200:
                return data.get("message","city not found")
            temp = data["main"]["temp"]
            condition = data["weather"][0]["main"]
            return {"city":city,"temp":temp,"condition":condition}
        except requests.exceptions.RequestException as e:
            return None,"API error, check Internet connection"
    def log_weather_data(self):
        city = input("Enter city name: ").strip()
        today = str(date.today())
        weather = self.get_weather_info(city)
        with open(self.log_file,"r") as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                if row["Date"] == today and row["City"].lower() == city.lower():
                    return f"Weather log for {city} was already logged today"
        with open(self.log_file,"a") as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow([today,weather["city"],weather["temp"],weather["condition"]])
        print(f"Weather log for {city} logged successfully")
    def view_all_logs(self):
        with open(self.log_file,"r") as csvfile:
            reader = csv.DictReader(csvfile)
            rows = list(reader)
            if not rows:
                return "No Weather logs found"
            print("\n{:<20}{:<20}{:<20}{:<20} ".format("Date","City","Temperature","Condition"))
            print("-"*75)
            for row in rows:
                print("{:<20}{:<20}{:<20}{:<20}".format(row["Date"],row["City"].title(),row["Temperature"],row["Condition"]))
```
