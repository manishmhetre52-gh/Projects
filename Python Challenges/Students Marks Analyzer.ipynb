```python
""" Challenge: Student Marks Analyzer

Create a Python program that allows a user to input student names along with their marks and then calculates useful statistics.
Your program should:
1. Let the user input multiple students with their marks (name + integer score).
2. After input is complete, display:
   - Average marks
   - Highest marks and student(s) who scored it
   - Lowest marks and student(s) who scored it
   - Total number of students

Bonus:
- Allow the user to enter all data first, then view the report
- Format output clearly in a report-style layout
- Prevent duplicate student names
"""
```

```python
class Student:
    def __init__(self,name,score):
        self.name = name
        self.score = score
class Analyzer:
    def __init__(self):
        self.student_list = []
    def add_student(self):
        while True:
            name = input("Enter student name('STOP' to stop): ").strip().upper()
            if name == "STOP":
                return None
            for student in self.student_list:
                if student.name == name:
                    return f"Student with name {name} already exists"
            score = int(input("Enter student marks score: "))
            if score < 0 or score > 100:
                return "Invalid score entered"
            self.student_list.append(Student(name,score))
            print(f"Student with name {name} added")
    def highest_marks(self):
        top_student = sorted(self.student_list, key=lambda student: student.score, reverse=True)[0]
        return f"Top student with highest marks {top_student.score} is {top_student.name}"
    def lowest_marks(self):
        bottom_student = sorted(self.student_list, key=lambda student: student.score)[0]
        return f"Top student with lowest marks {bottom_student.score} is {bottom_student.name}"
    def total_students(self):
        return f"Total number of students: {len(self.student_list)}"
    def average_marks(self):
        tot = 0
        for student in self.student_list:
            tot += student.score
        avg = tot / len(self.student_list)
        return f"Average marks: {avg}"
    def main(self):
        while True:
            menu = input("Choose your option:\n1.Add Student.\n2.Get Report.3.Exit").strip()
            if menu == "1":
                print(self.add_student())
            elif menu == "2":
                if self.student_list:
                    print(self.highest_marks())
                    print(self.lowest_marks())
                    print(self.total_students())
                    print(self.average_marks())
                else:
                    return f"No students in the records"
            elif menu == "3":
                break
            else:
                print("Invalid option")
```

```python
analyzer = Analyzer()
```

```python
analyzer.main()
```
