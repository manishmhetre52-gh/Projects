```python
class Hospital:
    def __init__(self,name,capacity):
        self.name = name.capitalize()
        self.capacity = capacity
        self.patients = []
    def view_patients(self):
        for patient in self.patients:
            print(f"Patient Name: {patient.name}, Patient Urgency Level: {patient.status_format()}")
class Patient:
    def __init__(self,name,urgency_level):
        self.name = name.upper()
        self.urgency_level = urgency_level
    def status_format(self):
        if self.urgency_level == 0:
            return "Normal"
        elif self.urgency_level == 1:
            return "Urgent"
        elif self.urgency_level == 2:
            return "Super Urgent"
        return None
    def __str__(self):
        return f"Patient Name: {self.name}, Patient Urgency Level: {self.status_format()}"
class Specialization:
    def __init__(self,hospital):
        self.hospital = hospital
    def add_patient(self,name,urgency_level):
        if len(self.hospital.patients) == self.hospital.capacity:
            return f"Sorry, Maximum queue capacity reached for {self.hospital.name} Hospital Queue"
        else:
            name = name.upper()
            for patient in self.hospital.patients:
                if patient.name == name:
                    return f"Patient {name} already exists in the Hospital Queue"
            temp_patient = Patient(name,urgency_level)
            if urgency_level == 2:
                self.hospital.patients.insert(0, temp_patient)
                return f"Patient {name},added to the Hospital Queue at high priority level"
            elif urgency_level == 1:
                index = 0
                while index < len(self.hospital.patients) and self.hospital.patients[index].urgency_level == 2:
                    index += 1
                self.hospital.patients.insert(index, temp_patient)
                return f"Patient {name} added to the Hospital Queue at a priority level"
            else:
                self.hospital.patients.append(temp_patient)
                return f"Patient {name} added to the Hospital Queue"
    def retrieve_next_patient(self):
        if self.hospital.patients:
            patient = self.hospital.patients.pop(0)
            return f"Patient Name: {patient.name}, Patient Urgency Level: {patient.status_format()}"
        else:
            return "No patients in the Hospital Queue"
    def remove_patient(self,name):
        if self.hospital.patients:
            for patient in self.hospital.patients:
                if patient.name == name.upper():
                    self.hospital.patients.remove(patient)
                    return f"Patient {name.upper()} removed from the Hospital Queue"
            return f"Sorry, Patient {name.upper()} does not exist in the Hospital Queue"
        else:
            return "No patients in the Hospital Queue"
    def check_queue_capacity(self):
        return f"{len(self.hospital.patients)} patients are in the {self.hospital.name} Hospital Queue, and maximum queue capacity for the hospital is {self.hospital.capacity}"
class OperataionManager:
    def __init__(self,hospital):
        self.hospital = hospital
    def add_patient(self):
        if len(self.hospital.patients) == self.hospital.capacity:
            return f"Sorry, Maximum queue capacity reached for {self.hospital.name} Hospital Queue"
        else:
            name = input("Enter patient Name: ").strip().upper()
            urgency_level = int(input("Enter urgency level(0,1,2): "))
            for patient in self.hospital.patients:
                if patient.name == name:
                    return f"Patient {name} already exists in the Hospital Queue"
            temp_patient = Patient(name,urgency_level)
            if urgency_level == 2:
                self.hospital.patients.insert(0, temp_patient)
                return f"Patient {name},added to the Hospital Queue at high priority level"
            elif urgency_level == 1:
                index = 0
                while index < len(self.hospital.patients) and self.hospital.patients[index].urgency_level == 2:
                    index += 1
                self.hospital.patients.insert(index, temp_patient)
                return f"Patient {name} added to the Hospital Queue at a priority level"
            else:
                self.hospital.patients.append(temp_patient)
                return f"Patient {name} added to the Hospital Queue"
    def list_all_patients(self):
        for patient in self.hospital.patients:
            print(f"Patient Name: {patient.name}, Patient Urgency Level: {patient.status_format()}")
    def list_patients_urgency_lvl(self):
        urgency_level = input("Enter urgency level(Normal, Urgent, Super Urgent): ) ").strip().title()
        found = False
        for patient in self.hospital.patients:
            if urgency_level == patient.status_format():
                print(f"Patient Name: {patient.name}, Patient Urgency Level: {patient.status_format()}")
                found = True
        if not found:
            print(f"No Patient with urgency level {urgency_level} in the hospital queue")
    def retrieve_next_patient(self):
        if self.hospital.patients:
            patient = self.hospital.patients.pop(0)
            return f"Patient Name: {patient.name}, Patient Urgency Level: {patient.status_format()}"
        else:
            return "No patients in the Hospital Queue"
    def remove_patient(self):
        name = input("Enter patient Name: ").strip().upper()
        for patient in self.hospital.patients:
            if patient.name == name:
                self.hospital.patients.remove(patient)
                return f"Patient {name} removed from the Hospital Queue"
        return f"Sorry, Patient {name} does not exist in the Hospital Queue"
    def main(self):
        while True:
            menu = input("Welcome to the Hospital Operation Management System.\nChoose your option.\n1.Add Patient\n2.List Patients according to urgency level\n3.List all Patients\n4.Retrieve next Patient\n5.Remove Patient\n6.Exit\n").strip()
            if menu == "1":
                print(self.add_patient())
            elif menu == "2":
                self.list_patients_urgency_lvl()
            elif menu == "3":
                self.list_all_patients()
            elif menu == "4":
                print(self.retrieve_next_patient())
            elif menu == "5":
                print(self.remove_patient())
            elif menu == "6":
                print("Thank you for using this Interface.\nLogging Out...")
                break
            else:
                print("Invalid Input")
```
