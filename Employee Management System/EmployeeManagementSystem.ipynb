```python
import logging
```

```python
logger = logging.getLogger(__name__)
logger.setLevel(logging.DEBUG)
```

```python
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
```

```python
info_handler = logging.FileHandler("EMS_INFO.log")
info_handler.setLevel(logging.INFO)
info_handler.setFormatter(formatter)
logger.addHandler(info_handler)
```

```python
class Company:
    def __init__(self,name):
        self.name = name
        self.employee_list = []
        self.manager_list = []
    def add_manager(self,manager):
        self.manager_list.append(manager)
        logger.info(f"{manager.man_id} was added in {self.name}")
        return f"{manager.man_id} added successfully"
    def view_managers(self):
        for manager in self.manager_list:
            print(f"Manager ID:{manager.man_id}")
        logger.info(f"Company {self.name} fetched Managers records")
    def view_employees(self):
        for employee in self.employee_list:
            print(f"Employee Name: {employee.name}, Employee Age: {employee.age}, Employee Salary: {employee.salary}")
        logger.info(f"Company {self.name} fetched Employees records")
class Employee:
    def __init__(self,name,age,salary):
        self.name = name
        self.age = age
        self.salary = salary
class EmployeeManager:
    def __init__(self,man_id,man_password,company):
        self.man_id = man_id
        self.man_password = man_password
        self.company = company
    def add_employee(self,name,age,salary):
        for emp in self.company.employee_list:
            if emp.name == name:
                logger.warning(f"Manager {self.man_id} tried to add an existing Employee {name} in the company records")
                return f"Employee {name} already exists in the records"
        emp = Employee(name,age,salary)
        self.company.employee_list.append(emp)
        logger.info(f"Manager {self.man_id} added Employee {name} in the company records")
        return f"Employee {name} added successfully"
    def remove_employee(self,name):
        for employee in self.company.employee_list:
            if employee.name == name:
                self.company.employee_list.remove(employee)
                logger.info(f"Manager {self.man_id} removed Employee {name} from the company records")
                return f"Employee {name} removed successfully"
        logger.warning(f"Manager {self.man_id} tried to remove non existing Employee with name:{name}, from the company records")
        return f"Employee {name} not found"
    def get_employee_list(self):
        for employee in self.company.employee_list:
            print(f"Employee Name: {employee.name}, Employee Age: {employee.age}, Employee Salary: {employee.salary}")
            logger.info(f"Manager {self.man_id} fetched {self.company} company employee list")
    def get_employee_by_name(self,name):
        for employee in self.company.employee_list:
            if employee.name == name:
                logger.info(f"Manager {self.man_id} fetched records for Employee {name}")
                return f"Employee Name: {employee.name}, Employee Age: {employee.age}, Employee Salary: {employee.salary}"
        logger.warning(f"Manager {self.man_id} tried to fetch records of non existing Employee with name:{name}")
        return f"Employee {name} not found"
    def change_emp_salary(self,name,salary):
        if salary <= 0:
            logger.error(f"{self.man_id} tried to input negative/zero salary for Employee {name}")
            return f"Salary cannot be negative or zero"
        for employee in self.company.employee_list:
            if employee.name == name:
                employee.salary = salary
                logger.info(f"Manager {self.man_id} changed salary for Employee {name} to {salary}")
                return f"Salary for employee {name} changed successfully to {salary}"
        logger.warning(f"{self.man_id} tried to change salary for a non-existing Employee with name:{name}")
        return f"Employee {name} not found"
    def del_employee_by_age_range(self,age_range):
        count = 0
        for employee in self.company.employee_list[:]:
            if age_range[0] <= employee.age <= age_range[1]:
                print(f"Employee {employee.name} removed successfully")
                logger.info(f"Manager {self.man_id} removed Employee {employee.name} from the {self.company} company records")
                self.company.employee_list.remove(employee)
                count += 1
        logger.info(f"Manager {self.man_id} removed {count} Employees within the age range {age_range}")
        return f"{count} Employees within the age range {age_range} removed successfully"
class FrontendManager:
    man_id = None
    def __init__(self,company):
        self.company = company
    def login(self):
        login_id = input("Enter employee id : ").strip().lower()
        password = input("Enter employee password : ").strip().lower()
        for manager in self.company.manager_list:
            if manager.man_id == login_id and manager.man_password == password:
                print(f"{manager.man_id} is logged in")
                logger.info(f"Manager {login_id} from {self.company.name} company, successfully logged in")
                self.man_id = login_id
                return True
        print(f"Invalid Credentials")
        logger.warning(f"Invalid Login attempt for id {login_id}")
        return False
    def add_employee(self):
        name = input("Enter employee name : ").strip().lower()
        age = int(input("Enter employee age : "))
        if age <= 0:
            logger.error(f"Manager {self.man_id} tried to input negative/zero age for Employee {name}")
            return f"Age cannot be negative/zero for employee"
        salary = int(input("Enter employee salary : "))
        if salary <= 0:
            logger.error(f"Manager {self.man_id} tried to input negative/zero salary for Employee {name}")
            return f"Salary cannot be negative/zero for employee"
        for employee in self.company.employee_list:
            if employee.name == name:
                logger.warning(f"Manager {self.man_id} tried to add an existing employee {name} to the company records")
                return f"Employee {name} already exists in the records"
        self.company.employee_list.append(Employee(name,age,salary))
        logger.info(f"Manager {self.man_id} added Employee {name} to the company records")
        return f"Employee {name} added successfully in the records"
    def remove_employee(self):
        name = input("Enter employee name : ").strip().lower()
        for employee in self.company.employee_list:
            if employee.name == name:
                self.company.employee_list.remove(employee)
                logger.info(f"Manager {self.man_id} removed Employee {name} from the company records")
                return f"Employee {name} removed successfully from the records"
        logger.warning(f"Manager {self.man_id} tried to remove non existing Employee with name:{name}, from the company records")
        return f"Employee {name} does not exist in the records"
    def get_employee_list(self):
        for employee in self.company.employee_list:
            print(f"Employee Name: {employee.name}, Employee Age: {employee.age}, Employee Salary: {employee.salary}")
        logger.info(f"Manager {self.man_id} fetched company employee records")
    def get_employee_by_name(self):
        name = input("Enter employee name : ").strip().lower()
        for employee in self.company.employee_list:
            if employee.name == name:
                logger.info(f"Manager {self.man_id} fetched records for Employee {name} from company records")
                return f"Employee Name: {employee.name}, Employee Age: {employee.age}, Employee Salary: {employee.salary}"
        logger.warning(f"Manager {self.man_id} tried to fetch records for non existing employee with name:{name}, from the company records")
        return f"Employee {name} does not exist in the records"
    def change_emp_salary(self):
        name = input("Enter employee name : ").strip().lower()
        for employee in self.company.employee_list:
            if employee.name == name:
                salary = int(input("Enter employee salary : "))
                if salary <= 0:
                    logger.error(f"Manager {self.man_id} tried to input negative/zero salary for Employee {name}")
                    return f"Salary cannot be negative/zero for employee"
                employee.salary = salary
                logger.info(f"Manager {self.man_id} changed salary for Employee {name} to {salary}")
                return f"Salary for employee {name} has been updated successfully"
        logger.warning(f"{self.man_id} tried to change salary for a non-existing Employee with name:{name}")
        return f"Employee {name} does not exist in the records"
    def del_employee_by_age_range(self):
        age_range = input("Enter employee age range (e.g 20-30) : ").split("-")
        if len(age_range)<2:
            logger.error(f"Manager {self.man_id} tried to input invalid age range input")
            return f"Invalid age range input"
        count = 0
        for employee in self.company.employee_list[:]:
            if int(age_range[0]) <= employee.age <= int(age_range[1]):
                print(f"Employee {employee.name} removed successfully from the records")
                logger.info(f"Manager {self.man_id} removed Employee {employee.name} from the company records")
                self.company.employee_list.remove(employee)
                count += 1
        logger.info(f"Manager {self.man_id} removed {count} Employees within the age range {age_range}")
        return f"{count} Employees within the range {age_range} removed successfully"
    def main(self):
        login = self.login()
        if login:
            while True:
                menu = input("Welcome to the Frontend Manager Interface\nChoose Your Option\n1.Add Employee\n2.Remove Employee\n3.Get Employee List\n4.Get Employee details\n5.Update Employee Salary\n6.Delete Employees within a range\n7.Exit")
                if menu == "1":
                    print(self.add_employee())
                elif menu == "2":
                    print(self.remove_employee())
                elif menu == "3":
                    self.get_employee_list()
                elif menu == "4":
                    print(self.get_employee_by_name())
                elif menu == "5":
                    print(self.change_emp_salary())
                elif menu == "6":
                    print(self.del_employee_by_age_range())
                elif menu == "7":
                    print("Thanks for using the Frontend Manager Interface. Logging Out")
                    logger.info(f"Manager {self.man_id} logged out successfully")
                    self.man_id = None
                    break
                else:
                    print("Invalid Input")
                    logger.error(f"Invalid Input from Manager {self.man_id}")
tata = Company("TATA")
m1 = EmployeeManager("manishmhetre","1234",tata)
m2 = EmployeeManager("viratkohli","1818",tata)
tata.add_manager(m1)
tata.add_manager(m2)
m1.add_employee("manish",26,100000)
m2.add_employee("virat",37,180000)
m2.add_employee("rohit",36,450000)
m2.add_employee("dhoni",42,700000)
print(tata.view_employees())
print(tata.view_managers())
```

```python
fm1 = FrontendManager(tata)
```
